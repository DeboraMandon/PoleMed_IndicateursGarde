```python
plages_horaires=['08h-09h', '09h-10h', '10h-11h', '11h-12h', '12h-13h', '13h-14h',
       '14h-15h', '15h-16h', '16h-17h', '17h-18h', '18h-19h', '19h-20h',
       '20h-21h', '21h-22h', '22h-23h', '23h-00h', '00h-01h', '01h-02h',
       '02h-03h', '03h-04h', '04h-05h', '05h-06h', '06h-07h', '07h-08h']

for plage in plages_horaires:
    df_planning[plage] = 0

# Vérification des conditions et mise à jour des colonnes
for index, row in df_planning.iterrows():
    heure_debut_str = row['Heure_Début']
    duree = row['Durée']
    heure_debut = int(heure_debut_str.split(':')[0])
    minute_debut = int(heure_debut_str.split(':')[1])

    for plage in plages_horaires:
        plage_debut = int(plage.split('-')[0].replace('h', ''))

        if plage_debut <= heure_debut < plage_debut + duree:
            df_planning.loc[index, plage] = 1

```
